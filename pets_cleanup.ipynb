```python
import pandas as pd
```

```python
path = "Resources/all_records.csv"
pets_df = pd.read_csv(path)
pets_df.head()
```

```python
for col in pets_df.columns: 
    print(col)
```

```python
# Filter dataframe for desired columns and rename as needed
pet_columns = ["Animal ID", "DateTime_intake", "Intake_Type", "IntakeCondition", "Animal_Type_intake", "Sex", "Age", "Breed_intake",
               "Color_intake", "DateTime_outcome", "Outcome_Type", "Age_upon_Outcome", "Age_Bucket", "retriever", "shepherd", "beagle", 
               "terrier", "boxer", "poodle", "rottweiler", "dachshund", "chihuahua", "pit bull", "DateTime_length", "Days_length"]

pets_transformed= pets_df[pet_columns].copy()

pets_transformed = pets_transformed.rename(columns={"IntakeCondition": "Condition_intake",
                                                    "Age": "Age_intake"})

pets_transformed.head()
```

```python
#Drop null values
pets_transformed = pets_transformed.dropna()
```

```python
pets_transformed["Animal_Type_intake"].unique()
```

```python
pets_transformed["Intake_Type"].unique()
```

```python
#Remove values for wildlife intake
wildlife = pets_transformed[ pets_transformed['Intake_Type'] == "Wildlife" ].index

pets_transformed.drop(wildlife , inplace=True)
```

```python
#Remove Other, Bird, Livestock to leave only cat and dog values
bird = pets_transformed[ pets_transformed['Animal_Type_intake'] == "Bird" ].index
other = pets_transformed[ pets_transformed['Animal_Type_intake'] == "Other" ].index
livestock = pets_transformed[ pets_transformed['Animal_Type_intake'] == "Livestock" ].index


pets_transformed.drop(bird , inplace=True)
pets_transformed.drop(other , inplace=True)
pets_transformed.drop(livestock , inplace=True)
```

```python
#Remove unknown values for sex
unknown = pets_transformed[ pets_transformed['Sex'] == "Unknown" ].index
pets_transformed.drop(unknown , inplace=True)
```
